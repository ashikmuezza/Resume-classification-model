```python
import fitz
import os
import re
import shutil
import pandas as pd
from docx2python import docx2python
from nltk.tokenize import word_tokenize
from nltk.tokenize.treebank import TreebankWordDetokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()


#shutil.rmtree('datasets')

# Creating directory for saving cleaned datasets

if not os.path.exists('datasets'):
  os.mkdir('datasets')


# Removing blank spaces & Changed words Capital to lower
# Tokenize the text file
# Stemming, and Removing stopwords 
# Detokenize to Strings

def clean(text):
  text = re.sub('([^A-Za-z0-9 ]|[^ ]*[0-9][^ ]*)', '', text.strip().lower())
  tokens = word_tokenize(text)
  removed_tokens = [ porter_stemmer.stem(word) for word in tokens if word not in stopwords.words()]
  merged_tokens = TreebankWordDetokenizer().detokenize(removed_tokens)
  return merged_tokens


# Reading resumes 

category = list()
resumes = list()

directory = 'resumes'
cnt = 0
for file in os.listdir(directory):
  cnt += 1
  path = '/content/resumes/'+file
  file = file.replace('_',' ')
  print(file)
  if '.pdf' in file:
    file = re.sub('[0-9]','',file.replace('.pdf','').strip())  
    save_path = 'datasets/'+file
    if not os.path.exists(save_path):
      os.mkdir(save_path)
    print(file)
  
    doc = fitz.open(path)
    text = ""
    for page in doc:
      text+=page.getText()

  elif '.docx' in file:
    file = re.sub('[0-9]','',file.replace('.docx','').strip())
    save_path = 'datasets/'+file
    if not os.path.exists(save_path):
      os.mkdir(save_path)
    print(file)
    doc_result = docx2python(path)
    text = doc_result.text

  cleaned_text = clean(text) 
  
  category.append(file)
  resumes.append(cleaned_text)


  lst = os.listdir(save_path) # your directory path
  count = len(lst) + 1
  text_file = open(save_path +"/"+ str(count) +".txt", "wt")
  text_file.write(cleaned_text)
  text_file.close()
  #if cnt > 10:
    #break


data = {"category":category[:-1],"resumes" : resumes[:-1] }
df = pd.DataFrame(data)
df.to_csv('final_data.csv')
```
